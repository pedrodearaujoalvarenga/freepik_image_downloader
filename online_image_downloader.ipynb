# Image Downloader

This scripts searches your search key in the Freepik website (https://www.freepik.com/) and download them.

```python
image_number = 350 #number of images to download
search_key = 'dessert with solid color background' #search key
folder_path = 'solid-color/desert' #folder will be created if does not exist
```

```python
import requests
from bs4 import BeautifulSoup
import os
from urllib.parse import urljoin
```

```python
search_key = search_key.replace(' ', '-')


images_downloaded = 0


original_link = 'https://www.freepik.com/free-photos-vectors/' + search_key

link_addition = ''
link_addition_number = 1


while(images_downloaded<image_number):

    link = original_link + link_addition


    # Define the link to open

    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.4567.89 Safari/537.36'
    }

    # Send a GET request to the link
    response = requests.get(link, headers=headers)
    response.raise_for_status()

    # Create a BeautifulSoup object to parse the HTML content
    soup = BeautifulSoup(response.content, 'html.parser')

    # Create a folder to save the downloaded images
    os.makedirs(folder_path, exist_ok=True)

    # Find all the image tags in the HTML
    image_tags = soup.find_all('img')

    # Download and save the images
    for i, img_tag in enumerate(image_tags):
        try:
            # Get the source URL of the image
            src = img_tag['src']

            # Construct the absolute URL if the source URL is relative
            abs_url = urljoin(link, src)

            # Send a GET request to the image URL
            img_response = requests.get(abs_url)
            img_response.raise_for_status()

            # Extract the filename from the URL
            filename = os.path.basename(abs_url)

            if filename[0].isdigit(): continue
            if filename.startswith('default'): continue
            if not filename.endswith('png') and not filename.endswith('jpg'): continue

            # Save the image in the "downloaded_images" folder
            path = os.path.join(folder_path, filename)
            with open(path, 'wb') as file:
                file.write(img_response.content)
            images_downloaded +=1

            if images_downloaded == image_number: break

            print(f"Downloaded image {images_downloaded}: {filename}")

        except requests.exceptions.RequestException as e:
            print(f"Error downloading image {images_downloaded}: {e}")
        except IOError as e:
            print(f"Error saving image {images_downloaded}: {e}")
    
    link_addition = '/' + str(link_addition_number)
    link_addition_number += 1

print("Download complete!")
```
